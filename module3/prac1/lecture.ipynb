#### Можно использовать в виде декоратора

````python
@torch.no_grad()
def foo():
    a = torch.rand(3, 5, requires_grad = True)
    b = torch.rand(3, 5, requires_grad = True)
    loss = torch.mean(a - b)

    print(f'{loss=}')
    return a, b
````

````python
a, b = foo()
````

````python
torch.mean(a - b) # есть grad_fn - можно считать градиенты
````

#### Сделаем то же самое с inference_mode

````python
@torch.inference_mode()
def foo():
    a = torch.rand(3, 5, requires_grad = True)
    b = torch.rand(3, 5, requires_grad = True)
    loss = torch.mean(a - b)

    print(f'{loss=}')
    return a, b
````

````python
a, b = foo()
````

````python
torch.mean(a - b) # нет grad_fn - считать градиенты нельзя
````

#### ! Что можно обернуть в inference_mode - функция предсказания, оценки качества (не обучаем - не считаем градиенты)

### **Градиентный спуск своими руками**

````python
n_features = 2
n_objects = 300

torch.manual_seed(0)

w_true = torch.rand(n_features)
b_true = torch.rand(1)

x = (torch.rand(n_objects, n_features) - 0.5) * 10 * (torch.arange(n_features) * 2 + 1)
y = torch.matmul(x, w_true) + torch.rand(n_objects) + b_true
````

````python
n_steps = 200
step_size = 1e-2
````

````python
w = torch.rand(n_features, requires_grad = True)
b = torch.rand(1, requires_grad = True)

for i in range(n_steps):
    y_pred = torch.matmul(x, w) + b

    mse = torch.mean((y_pred - y) ** 2)

    if i<20 or i%10 == 0:
        print(f'MSE на шаге {i} равен: {mse:.5f}')

    mse.backward()

    with torch.no_grad():
        w -= w.grad * step_size
        b -= b.grad * step_size

    w.grad.zero_()
    b.grad.zero_()
````

````python
layer = nn.Linear(in_features = n_features, out_features = 1)
for i in range(n_steps):
    y_pred = layer(x)

    mse = torch.mean((y_pred - y) ** 2)

    if i<20 or i%10 == 0:
        print(f'MSE на шаге {i} равен: {mse:.5f}')

    mse.backward()

    with torch.no_grad():
        layer.weight -= layer.weight.grad * step_size
        layer.bias -= layer.bias.grad * step_size
    
    #layer.weight.grad.zero_()
    #layer.bias.grad.zero_()

    layer.zero_grad()
````

#### Сделаем 2 полносвязных слоя с функцией активации ReLU

````python
n_steps = 2000
step_size = 1e-3
````

````python
layer1 = nn.Linear(in_features = n_features, out_features = 3)
layer2 = nn.Linear(in_features = 3, out_features = 1)
activation = nn.ReLU()

for i in range(n_steps):
    y_pred = layer2(activation(layer1(x))).ravel()

    mse = torch.mean((y_pred - y) ** 2)

    if i < 20 or i % 50 == 0:
        print(f'MSE на шаге {i} равен: {mse:.5f}')

    mse.backward()

    with torch.no_grad():
        layer1.weight -= layer1.weight.grad * step_size
        layer1.bias -= layer1.bias.grad * step_size
        layer2.weight -= layer2.weight.grad * step_size
        layer2.bias -= layer2.bias.grad * step_size

    layer1.zero_grad()
    layer2.zero_grad()
    
````
