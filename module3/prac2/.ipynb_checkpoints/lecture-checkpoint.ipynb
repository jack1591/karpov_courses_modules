````python
input_tensor = torch.rand(10, 700)
````

````python
class CustomTaskNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.linear1 = nn.Linear(700, 500)
        self.linear2 = nn.Linear(500, 200)
        self.linear3 = nn.Linear(200, 10)

        self.activation = nn.ReLU()

    def forward(self, x):
        output = self.activation(self.linear1(x))
        output = self.activation(self.linear2(output))
        output = self.linear3(output)

        return output
````

### Датасеты и даталоадеры

````python
from torch.utils.data import Dataset, TensorDataset
````

````python
n_features = 2
n_objects = 300

torch.manual_seed(0)
````

````python
w_true = torch.randn(n_features)

X = (torch.rand(n_objects, n_features) - 0.5) * 10
X *= (torch.arange(n_features) * 3 + 2)
Y = (X @ w_true + torch.randn(n_objects)).unsqueeze(1)
````

````python
dataset = TensorDataset(X, Y)
````

````python
list(dataset)
````

````python
# Как выглядит TensorDataset

class CustomDataset(Dataset):
    def __init__(self, w_true, n_features, n_objects):
        self.X = (torch.rand(n_objects, n_features) - 0.5) * 10
        self.X *= (torch.arange(n_features) * 3 + 2)
        self.Y = (self.X @ w_true + torch.randn(n_objects)).unsqueeze(1)

    def __len__(self):
        return len(self.Y)

    def __getitem__(self, item):
        return self.X[item], self.Y[item]
````

````python
dataset = CustomDataset(w_true, n_features, n_objects)
````

````python
dataset[0], len(dataset)
````

### Обучение первой нейросети в ```PyTorch```

````python
class CustomTaskNetwork(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear = nn.Linear(n_features, 1)

    def forward(self, x):
        return self.linear(x)

net = CustomTaskNetwork()

optimizer = optim.Adam(net.parameters(), lr = 1e-1)

loss_fn = nn.MSELoss()

dataset = TensorDataset(X, Y)

loader = DataLoader(dataset, batch_size = 4, shuffle = True)

w_list = torch.empty(len(loader) + 1, n_features)
````

````python
net.train()

for i, (x, y) in enumerate(loader):
    w_list[i] = net.linear.weight.detach().clone() # запишем список весов для дальнейшей визуализации

    optimizer.zero_grad()

    output = net(x)

    loss = loss_fn(output, y)

    print(f'MSE на шаге {i} равен {loss.item():.5f}')
    
    loss.backward()

    optimizer.step()

w_list[len(loader)] = net.linear.weight.detach().clone()
````

````python
from plot_trajectory import plot_trajectory

plot_trajectory(
    w_list.numpy(),
    w_true.numpy(),
    X.numpy(),
    Y.squeeze(1).numpy(),
    'Mini-batch градиентный спуск'
)
````

````python
num_epochs = 5

net = CustomTaskNetwork()

optimizer = optim.Adam(net.parameters(), lr = 1e-1)

w_list = torch.empty(len(loader) * num_epochs + 1, n_features)
````

````python
net.train()

for epoch in range(num_epochs):
    
    total_loss = 0
    
    for i, (x, y) in enumerate(loader):
        w_list[i + epoch * len(loader)] = net.linear.weight.detach().clone() # запишем список весов для дальнейшей визуализации
    
        optimizer.zero_grad()
    
        output = net(x)
    
        loss = loss_fn(output, y)
    
        total_loss += loss.item()
        
        loss.backward()
    
        optimizer.step()
    
    total_loss /= len(loader)

    print(f'Средняя MSE на {epoch} эпохе: {total_loss:.5f}')
w_list[len(loader) * num_epochs] = net.linear.weight.detach().clone()
````
