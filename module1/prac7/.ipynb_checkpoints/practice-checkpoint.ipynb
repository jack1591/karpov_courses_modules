```python
class Triangle:
    a = 0
    b = 0
    c = 0
    n_dots = 3
    
    def __init__(self,a,b,c):
        if (a+b+c>2*max(a,b,c)):
            self.a,self.b,self.c = a,b,c
        else: raise ValueError("triangle inequality does not hold")

    def area(self):
        p = (self.a+self.b+self.c)/2
        return (p*(p-self.a)*(p-self.b)*(p-self.c))**0.5

    
        
tr_1 = Triangle(2,3,4)
tr_2 = Triangle(3,4,5)
square_1 = tr_1.area()
square_2 = tr_2.area()
square_2
# tr_3 = Triangle(3,4,10)
```

```python
class Triangle(BaseFigure):
    a = 0
    b = 0
    c = 0
    n_dots = 3
    
    def __init__(self,a,b,c):
        self.a, self.b, self.c = a, b, c
        super().__init__()
            
    def area(self):
        p = (self.a+self.b+self.c)/2
        return (p*(p-self.a)*(p-self.b)*(p-self.c))**0.5

    def validate(self):
        if (self.a+self.b+self.c>2*max(self.a,self.b,self.c)):
            return self.a, self.b, self.c
        else: raise ValueError("triangle inequality does not hold")
        


class Rectangle(BaseFigure):
    n_dots = 4
    a = 0
    b = 0

    def __init__(self, a, b):
        self.a, self.b = a, b
        super().__init__()

    def area(self):
        return self.a * self.b

    def validate(self):
        return self.a, self.b

tr1 = Triangle(3,4,5)
tr1.validate()
rect1 = Rectangle(3,4)
rect1.area()
```

```python
class Vector:
    coords = []
    def __init__(self, coords):
        self.coords = coords

    def __add__(self, other):
        if (len(self.coords)!=len(other.coords)):
            raise ValueError(f'left and right lengths differ: {len(self.coords)} != {len(other.coords)}')
        return Vector([self.coords[i]+other.coords[i] for i in range(len(self.coords))])

    def __str__(self):
        return f'{self.coords}'

    def __mul__(self,other):
        if (isinstance(other,Vector)):
            if (len(self.coords)!=len(other.coords)):
                raise ValueError(f'left and right lengths differ: {len(self.coords)} != {len(other.coords)}')
            return sum(self.coords[i]*other.coords[i] for i in range(len(self.coords)))
        else:
            return Vector([self.coords[i]*other for i in range(len(self.coords))])

    def __eq__(self, other):
        if (len(self.coords)!=len(other.coords)):
            return False
        for i in range(len(self.coords)):
            if (self.coords[i] != other.coords[i]):
                return False
        return True

    def __abs__(self):
        return sum(elem**2 for elem in self.coords)**0.5
            
vec1 = Vector([-12,5])
vec2 = Vector([1,2,3])
print(abs(vec1))
```

```python
class ParsesCookies:
    def cookies(self):
        return self.request['cookies']
    def is_authed(self):
        return 'auth_key' in self.request['cookies']

class ParsesBody():
    def body(self):
        return self.request['body']

class ParsesHeaders():
    def headers(self):
        return self.request['headers']
    def need_json(self):
        return 'content-type' in self.request['headers'] and self.request['headers']['content-type'] == 'application/json'
```

```python
import json
class JsonHandler(ParsesBody, ParsesHeaders):
    def __init__(self, request):
        self.request = request

    def process(self):
        if (self.need_json() == False):
            return None
        try:
            data = json.loads(self.body())
            return len(data)
        except:
            None
```

```python
class SecureTextHandler(ParsesBody, ParsesCookies):
    def __init__(self, request):
        self.request = request

    def process(self):
        if (not(self.is_authed())):
            return None
        return len(self.body())
```
