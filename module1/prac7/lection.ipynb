## Магические методы

```python
class A:
    field = 0
    
    def __init__(self, field):
        self.field = field
    
    def __add__(self, other):
        return A(self.field+other.field)

    def __str__(self):
        return f'this is example of class A with value: {self.field}'

    def __call__(self, field2):
        return f'the sum is: {self.field+field2}'
        
obj1 = A(1)
obj2 = A(2)
obj3 = obj1+obj2
print(obj3)
obj3(4)
```

```python
class Addable:
    def __add__(self, b: int):
        return self.a+b

class Printable:
    def __str__(self):
        return f'The value of a is: {self.a}'

class ChildClass(Addable, Printable):
    def __init__(self,a : int):
        self.a = a

child = ChildClass(3)
# print(Addable()+3) 'Addable' object has no attribute 'a'
# print(Printable()) 'Printable' object has no attribute 'a'
print(child)
print(child+7)
```
