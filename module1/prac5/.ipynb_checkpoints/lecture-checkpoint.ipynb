```python
df.head()
df_copy = df.copy().set_index('id')
df_copy.loc['AVwc__kCByjofQCxkdm-', 'city'] = 'Rybinsk'
df_copy.loc['AVwc__kCByjofQCxkdm-']
```

```python
#df_copy.reset_index(inplace = True)
df_copy.groupby('city', as_index = False).agg({'longitude': 'max'})
```

```python
df_copy.groupby(['city', 'longitude'], as_index = False)['latitude'].mean()
```

```python
df_copy['dateAdded'] = pd.to_datetime(df_copy['dateAdded'])
by_month = df_copy.groupby(
    df_copy['dateAdded'].dt.month
).agg({
    'city': 'count',
})

by_month.sample(10, random_state = 42)
```

```python
by_month.plot(
    grid = True,
    figsize = (16,9)
)
```
