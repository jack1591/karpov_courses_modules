```python
df_copy = data.copy()
df_copy['update_delta'] = (pd.to_datetime(df_copy['dateUpdated']) - pd.to_datetime(df_copy['dateAdded'])).dt.days
df_copy.describe()
```

```python
df = data.copy()
df['categories'] = df['categories'].fillna('')
df['categories_count'] = df['categories'].str.split(',').apply(lambda x: len(x) if x != [''] else 0)
df_copy = df[
    df['categories_count'] > 20
].groupby('province').agg({'longitude' : 'min'})
df_copy = df_copy.round(3)
df_copy.to_csv('prac19_4.csv', sep = ';')
```
