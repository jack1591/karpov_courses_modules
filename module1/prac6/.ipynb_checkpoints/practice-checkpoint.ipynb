```python
df_users = pd.read_sql(
    """
    SELECT *
    FROM "user"
    WHERE country != 'Russia' AND ((exp_group!=0 AND exp_group!=3) OR city = 'Minsk')
    """,
    con = con_uri
)
print([int(df_users.iloc[i]['id']) for i in range(4)])
```

```python
df_users = pd.read_sql(
    """
    SELECT country, AVG(age)
    FROM "user"
    GROUP BY country
    """,
    con = con_uri
)
df_users['avg'] = df_users['avg'].round(2)
cyprus_users_age = df_users[
    df_users['country'] == 'Cyprus'
]['avg']
float(cyprus_users_age.iloc[0])
```

```python
df_users = pd.read_sql(
    """
    SELECT exp_group, os, COUNT(id) AS total_users, MAX(age) AS max_age, MIN(age) AS min_age
    FROM "user"
    GROUP BY exp_group, os
    """,
    con=con_uri
)
df_users.to_csv("group_users_prac6.csv", sep = ',', index = False)
```

```python
df_posts = pd.read_sql(
    """
    SELECT topic, MAX(LENGTH(text)) AS message_length
    FROM "post"
    GROUP BY topic
    ORDER BY message_length DESC
    """,
    con = con_uri
)
df_posts[
    df_posts['message_length']>25000
].iloc[0]['topic']
```

```python
df_users = pd.read_sql(
    """
    SELECT country, COUNT(country)
    FROM "user"
    GROUP BY country
    ORDER BY count
    """,
    con = con_uri
)
df_users[
    df_users['count'] > 1000
].iloc[0]['country']
```

```python
df_users = pd.read_sql(
    """
    SELECT exp_group, AVG(age) as avg_age, COUNT(id) AS number_of_users
    FROM "user"
    WHERE city = 'Moscow'
    GROUP BY exp_group
    HAVING AVG(age) > 27.2
    """,
    con = con_uri
)
for i in range(df_users.shape[0]):
    print(int(df_users.iloc[i]['number_of_users']),end = ' ')

# print(f"{' '.join(int(df_users.iloc[i]['number_of_users']) for i in range(df_users.shape[0]))}")
# print([int(df_users.iloc[i]['number_of_users']) for i in range(df_users.shape[0])])
```

```python
df_join1 = pd.read_sql(
    """
    SELECT f.post_id, f.time, u.age, u.os
    FROM "feed_action" f
    JOIN "post" p ON f.post_id = p.id
    JOIN "user" u ON f.user_id = u.id
    WHERE u.city = 'Omsk' AND f.action = 'like'
    ORDER BY f.time DESC
    LIMIT 100
    """,
    con = con_uri
)

df_join1.to_csv('prac6_join1.csv', index = False, sep = ',')
```

```python
df_join2 = pd.read_sql(
    """
    SELECT u.city, COUNT(u.id)
    FROM "feed_action" f
    JOIN "user" u ON f.user_id = u.id
    JOIN "post" p ON f.post_id = p.id
    WHERE u.age = 36 AND f.action = 'view' AND p.topic = 'covid' AND DATE(f.time) = '2021-12-01'
    GROUP BY u.city
    ORDER BY count
    """,
    con = con_uri
)
df_join2.iloc[0]['city']
```

```python
my_df = pd.read_sql(
    """
    SELECT p.id, COUNT(f.action), MAX(f.time)
    FROM "post" p
    JOIN "feed_action" f ON p.id = f.post_id
    WHERE f.action = 'like'
    GROUP BY p.id
    ORDER BY count DESC
    """,
    con = con_uri
)
for i in range(3):
    print(my_df.iloc[i]['id'],end = ' ')
```
