```python
import pandas as pd
```

```python
data = pd.read_csv('ks_crashed.csv')
data.head()
```

```python
(data.shape[0] - data.dropna().shape[0])*100/data.shape[0]
```

```python
data.isna().sum()
```

```python
most_popular_category = data['Главная категория'].value_counts().index[0]
data['Главная категория'] = data['Главная категория'].fillna(most_popular_category)
```

```python
data['Валюта'] = data['Валюта'].fillna('Неизвестная валюта')
```

```python
data.isna().sum()
```

```python
grouped_means = data.groupby('Главная категория')['Цель в долларах'].transform("mean")
```

```python
data['Цель в долларах'] = data['Цель в долларах'].fillna(grouped_means)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
fig.set_size_inches(10,8)

data[data['Собрано в долларах']<100000]['Собрано в долларах'].hist(bins = 100)

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,8)

plt.boxplot(data['Собрано в долларах'].values)
plt.ylim(0,50000)
plt.show()
```

```python
q_low = data['Собрано в долларах'].quantile(0.25)
q_high = data['Собрано в долларах'].quantile(0.75)

tentacle_length = q_high - q_low
tentacle_length
```

```python
upper_tentacle = q_high + 3 * tentacle_length
```

```python
upper_tentacle
```

```python
data[data['Собрано в долларах'] < upper_tentacle]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
data = data.dropna()

tfidf = TfidfVectorizer()

tfidf.fit(data['Название'])
```

```python
first_doc = data['Название'][0]
first_doc
```

```python
tfidf.transform([first_doc])
tfidf.transform([first_doc]).todense()
```

```python
tfidf.get_feature_names_out()
```

```python
df = pd.DataFrame(tfidf.transform([first_doc]).T.todense(),
                 index = tfidf.get_feature_names_out(),
                 columns = ['tfidf'])
df.sort_values('tfidf', ascending = False).head(10)
```
