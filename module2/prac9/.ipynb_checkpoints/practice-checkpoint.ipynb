```python
data[numeric_columns].corr()
```

```python
def get_redundant_pairs(df):
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0,i+1):
            pairs_to_drop.add((cols[i],cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    cols_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    cols_corr = cols_corr.drop(labels = labels_to_drop).sort_values(ascending = False)
    return cols_corr[0:n]

print('Top Absolute Correlations')
print(get_top_abs_correlations(data[numeric_columns], n=50))
```

```python
def correlation(dataset, cols, threshold = 0.9):
    cols_corr = set()
    corr_matrix = dataset[cols].corr().abs()
    n = corr_matrix.shape[0]
    for i in range(n):
        for j in range(i):
            if corr_matrix.iloc[i,j]>=threshold and corr_matrix[j] not in cols_corr:
                col_name = corr_matrix[j]
                cols_corr.add(col_name)
                if col_name in dataset.columns:
                    del dataset[col_name]

correlation(data,numeric_columns,0.9)
```

# Обработка категориальных признаков

```python
categorical_columns = data.loc[:,data.dtypes == object].columns
categorical_columns
```

```python
new_data = data.copy()
```

```python
from sklearn.model_selection import KFold
data = data.drop(['log_salary_in_usd'], axis = 1)
for category in data.columns:
    if data[category].nunique()<7:
        one_hot = pd.get_dummies(new_data[category], prefix = category, drop_first = True)
        new_data = pd.concat((new_data.drop(category,axis = 1), one_hot), axis = 1) * 1
    else:
        category_data = new_data.copy()
        splitter = KFold(n_splits = 4, shuffle = True, random_state = 33)
        global_mean = new_data['log_salary_in_usd'].mean()
        for train_index, test_index in splitter.split(new_data):
            mean_target = new_data.iloc[train_index].groupby(category)['log_salary_in_usd'].mean()

            mean_values = category_data.loc[category_data.index.isin(test_index),category].map(mean_target)
            mean_values = mean_values.fillna(global_mean)
            category_data.loc[category_data.index.isin(test_index), category] = mean_values
        
        new_data = category_data
        new_data[category] = new_data[category].astype('float64')
data = new_data
```

```python
data
```

# Построение модели

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
X = data.drop(['log_salary_in_usd'], axis = 1)
Y = data['log_salary_in_usd']
splitter = KFold(n_splits = 4)
```

## Первая модель (без регуляризации)

```python
train_errors = []
test_errors = []

scaler = StandardScaler()
scaler.fit(X.values)

for train_index, test_index in splitter.split(X):
    X_train, X_test = X.values[train_index], X.values[test_index]
    Y_train, Y_test = Y.values[train_index], Y.values[test_index]

    model = LinearRegression()
    X_train_transformed = scaler.transform(X_train)
    X_test_transformed = scaler.transform(X_test)
    model.fit(X_train_transformed, Y_train)

    train_errors.append(np.mean((model.predict(X_train_transformed) - Y_train)**2))
    test_errors.append(np.mean((model.predict(X_test_transformed) - Y_test)**2))
    
print(f"MSLE на тренировочной выборке: {round(np.mean(train_errors),5)}")
print(f"MSLE на тестовой выборке: {round(np.mean(test_errors),5)}")
```

## Вторая модель (Ridge-регуляризация (L2))

```python
alphas = np.linspace(start = 0.001, stop = 1, num = 40)
alphas
```

```python
ridge_pipe = Pipeline([('scaler', StandardScaler()),
                      ('Ridge', Ridge(max_iter = 100000))])

param_grid = {
    "Ridge__alpha": alphas
}

search = GridSearchCV(ridge_pipe, param_grid,
                     cv = splitter, scoring = 'neg_mean_squared_error')

search.fit(X, Y)

print(f"CV score = {search.best_score_:.5f}")
print(search.best_params_)
```

```python
ridge_pipe.set_params(Ridge__alpha =  search.best_params_["Ridge__alpha"])

cv_result = cross_validate(ridge_pipe, X, Y,
                           cv = splitter, scoring = 'neg_mean_squared_error', return_train_score = True)

print(f"MSLE на тренировочной выборке: {-round(np.mean(cv_result['train_score']),5)}")
print(f"MSLE на тестовой выборке: {-round(np.mean(cv_result['test_score']),5)}")
```

## Третья модель (с Lasso-регуляризацией (L1))

```python
lasso_pipe = Pipeline([('scaler', StandardScaler()),
                      ('Lasso', Lasso(max_iter=100000))])

param_grid = {
    'Lasso__alpha': alphas
}

search = GridSearchCV(lasso_pipe,param_grid,
                      cv = splitter,scoring = 'neg_mean_squared_error')

search.fit(X, Y)

print(search.best_params_)
print(f"CV Score = {search.best_score_:.5f}")
```

```python
lasso_pipe.set_params(Lasso__alpha = search.best_params_['Lasso__alpha'])

cv_result = cross_validate(ridge_pipe, X, Y,
                           cv = splitter,scoring = 'neg_mean_squared_error',
                          return_train_score = True)

print(f"MSLE на тренировочной выборке: {-round(np.mean(cv_result['train_score']),5)}")
print(f"MSLE на тестовой выборке: {-round(np.mean(cv_result['test_score']),5)}")
```
