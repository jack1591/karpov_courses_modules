```python
def func(x):
    return x**3 - 6*x**2 - 15*x + 10

x = np.linspace(-5, 10, 150)
```

```python
fig = plt.figure()
fig.set_size_inches(10, 8)

plt.plot(x, func(x))

plt.xlabel('x')
plt.ylabel('f')
plt.show()
```

```python
def func_derivative(x):
    return 3*x**2 - 12*x - 15
```

```python
def descent_and_plot(X_start=10, learning_rate=0.01, threshold=0.000001):
    X_start = X_start
    learning_rate = learning_rate
    threshold = threshold

    iterations = 1
    X_next = X_start - learning_rate*func_derivative(X_start)
    fig = plt.figure()
    fig.set_size_inches(10, 8)

    plt.scatter([X_start, X_next],
               [func(X_start), func(X_next)],
               color = 'green',
               s = 50)

    plt.plot([X_start, X_next],
               [func(X_start), func(X_next)],
               color = 'red')
        
    plt.xlabel('x')
    plt.ylabel('f')

    plt.plot(x, func(x), alpha = 0.5)

    while abs(X_next - X_start) > threshold:
        
        X_start = X_next
        X_next = X_start - learning_rate*func_derivative(X_start)

        plt.scatter([X_start, X_next],
               [func(X_start), func(X_next)],
               color = 'green',
               s = 50)

        plt.plot([X_start, X_next],
                   [func(X_start), func(X_next)],
                   color = 'red')
        iterations+=1
        
    print(f'     : X = {X_next} Y ={func(X_next)}')
    print(f'   : {iterations}')
    plt.show()
```

```python
descent_and_plot(X_start = 10, learning_rate = 0.01, threshold = 0.0001)
```
