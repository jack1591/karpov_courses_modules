```python
import pandas as pd
data = pd.read_csv('ks.csv')
data = data[data['Состояние'].isin(['successful', 'failed'])]
```

```python
import matplotlib.pyplot as plt
main_category = (data.groupby('Главная категория')['Собрано в долларах'].mean()).sort_values()
```

```python
fig = plt.figure()
fig.set_size_inches(10,8)

plt.bar(main_category.index, main_category)
plt.xticks(rotation=45)

plt.show()
```

```python
main_category_count = main_category = (data.groupby('Главная категория')['Собрано в долларах'].count())
fig = plt.figure()
fig.set_size_inches(10,8)

plt.bar(main_category_count.index, main_category_count)
plt.xticks(rotation=45)

plt.show()
```

```python
currency_dict = {}
for currency in data['Валюта'].unique():
    currency_dict[currency] = data[data['Валюта'] == currency]['Собрано в долларах']
```

```python
fig = plt.figure()
fig.set_size_inches(10,8)

plt.boxplot(currency_dict.values(),
           tick_labels = currency_dict.keys())

plt.ylim(0,50000)

plt.show()
```
