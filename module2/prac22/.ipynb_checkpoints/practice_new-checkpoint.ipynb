## Загрузим исходные данные

```python
engine = create_engine(
    "postgresql://robot-startml-ro:pheiph0hahj1Vaif@"
    "postgres.lab.karpov.courses:6432/startml"
)

user_query = "SELECT * FROM public.user_data"
user_data = pd.read_sql(user_query, engine)
user_data
```

```python
post_query = "SELECT * FROM public.post_text_df"
post_data = pd.read_sql(post_query, engine)
post_data
```

```python
feed_query = "SELECT * FROM public.feed_data LIMIT 300000"
feed_data = pd.read_sql(feed_query, engine)
feed_data
```

## Подготивим данные к обучению

```python
# запись с лайками дублирует запись с просмотром, где таргет равен 1
# это означает, что после просмотра поста, пользователь лайкнул его
feed_data = feed_data[feed_data['action']!='like']
feed_data
```

```python
# мы должны уметь предсказывать рекомендации для каждого пользователя
# значит, нам нужно оставить все user_id

df = pd.merge(
    user_data,
    feed_data,
    on = 'user_id',
    how = 'left'
)
df
```

```python
df = pd.merge(
    df,
    post_data,
    on = 'post_id',
    how = 'left'
)
df
```

```python
# заполним пропуски
df['timestamp'] = df['timestamp'].fillna(method = 'ffill').fillna(method = 'bfill')

most_popular_topic = df.describe(include = object)['topic'].loc['top']
df['topic'] = df['topic'].fillna(most_popular_topic)
df['target'] = df['target'].fillna(0).astype(int)

df
```

```python
# разобьем дату и удалим некоторые колонки - 'text', 'action', 'post_id'
df['hour'] = pd.to_datetime(df['timestamp']).apply(lambda x: x.hour)
df['day'] = pd.to_datetime(df['timestamp']).apply(lambda x: x.day)
df['month'] = pd.to_datetime(df['timestamp']).apply(lambda x: x.month)
df['post_id'] = df['post_id'].fillna(-1).astype(int)
df = df.drop(['text', 'action'], axis = 1)
```

```python
df
```

```python
# получили датафрейм без пропусков
df.isna().sum()
```

## Построение модели

```python
# разделим данные на тренировочную и тестовую части по дате, 
# потому что тестирование будет на тех же пользователях, но в будущем

df_train = df[df.timestamp < '2021-12-15']
df_test = df[df.timestamp >= '2021-12-15']

# удалим дату, потому что мы её уже разбили
df_train = df_train.drop('timestamp', axis = 1)
df_test = df_test.drop('timestamp', axis = 1)

X_train = df_train.drop('target', axis = 1)
y_train = df_train['target']

X_test = df_test.drop('target', axis = 1)
y_test = df_test['target']
```

```python
object_cols = df.loc[:, df.dtypes == object].columns.tolist()
object_cols
```

```python
from catboost import CatBoostClassifier

catboost = CatBoostClassifier(iterations = 500,
                             max_depth = 2,
                             learning_rate = 1)
catboost.fit(X_train, y_train, object_cols)
```

```python
from sklearn.metrics import roc_auc_score
print(f'Качество на тренировочной выборке roc_auc равно {roc_auc_score(y_train, catboost.predict_proba(X_train)[:, 1])}')
print(f'Качество на тестовой выборке roc_auc равно {roc_auc_score(y_test, catboost.predict_proba(X_test)[:, 1])}')
```

## Выполним преобразования данных для дальнейшего отбора M объектов

```python
# Создадим словарь, где для каждого топика будет храниться список id постов по этому топику
sorted_posts = feed_data.groupby('post_id').sum('target').sort_values('target', ascending = False)
sorted_posts = pd.merge(sorted_posts,
                       post_data,
                       on = 'post_id',
                       how = 'left')
sorted_posts = sorted_posts[['post_id', 'topic']]
topics = sorted_posts['topic'].unique()
posts_by_topic = {}
for topic in topics:
    posts_by_topic[topic] = []
    
for post in sorted_posts.values:
    posts_by_topic[post[1]].append(post[0])
```

```python
topics.shape
```

```python
post_topics = pd.merge(
    feed_data,
    post_data,
    on = 'post_id',
    how = 'left'
)
post_topics = post_topics[post_topics['target'] == 1]
post_topics = post_topics[['user_id', 'topic']]
post_topics
```

```python
liked_topics_by_user = {}
for post in post_topics.values:
    user = post[0]
    post_topic = post[1]
    if user not in liked_topics_by_user:
        liked_topics_by_user[user] = {}
        for topic in topics:
            liked_topics_by_user[user][topic] = 1
    liked_topics_by_user[user][post_topic]+=1
```

```python
def get_top_recommendations(user_id, users_df, posts_df, model, limit):
    POST_COUNT = 1400
    user = users_df[users_df['user_id'] == user_id].iloc[0]
    user_id = user['user_id']
    if user_id not in liked_topics_by_user:
        liked_topics_by_user[user_id] = {}
        for topic in topics:
            liked_topics_by_user[user_id][topic] = 1
    parts_count = 0
    for topic in topics:
        parts_count+=liked_topics_by_user[user_id][topic]

    all_posts = []

    for topic in topics:
        num_ids = liked_topics_by_user[user_id][topic] * POST_COUNT // parts_count
        for i in range(min(len(posts_by_topic[topic]),num_ids)):
            all_posts.append(posts_by_topic[topic][i])

    candidate_features = []
    for id_post in all_posts:
        post = posts_df[posts_df['post_id'] == id_post].iloc[0]
        features = {
            'user_id': user['user_id'],
            'gender': user['gender'],
            'age': user['age'],
            'country': user['country'],
            'city': user['city'],
            'exp_group': user['exp_group'],
            'os': user['os'],
            'source': user['source'],
            'post_id': post['post_id'],
            'topic': post['topic'],
            'hour': 21,
            'day': 7,
            'month': 11
        }
        candidate_features.append(features)
        
    candidates_df = pd.DataFrame(candidate_features)
    probabilities = np.array(model.predict_proba(candidates_df)[:, 1])
    
    top_indices = np.argpartition(probabilities, -limit)[-limit:]
    sorted_top_indices = top_indices[np.argsort(probabilities[top_indices])[::-1]]
    
    top_post_ids = [all_posts[ind] for ind in sorted_top_indices.tolist()]
    
    top_posts = posts_df[posts_df['post_id'].isin(top_post_ids)]
    return top_posts
```

```python
user_id = 10000
top_recommendations = get_top_recommendations(user_id, user_data, post_data, catboost, 5)
type(top_recommendations)
```

```python
import pandas as pd
from sqlalchemy import create_engine

def load_features() -> pd.DataFrame:
    df = batch_load_sql('SELECT * FROM bezyazychnyy_artem_features_lesson_22')
    return df
    
def batch_load_sql(query: str) -> pd.DataFrame:
    CHUNKSIZE = 200000
    engine = create_engine(
        "postgresql://robot-startml-ro:pheiph0hahj1Vaif@"
        "postgres.lab.karpov.courses:6432/startml"
    )
    conn = engine.connect().execution_options(stream_results=True)
    chunks = []
    for chunk_dataframe in pd.read_sql(query, conn, chunksize=CHUNKSIZE):
        chunks.append(chunk_dataframe)
    conn.close()
    return pd.concat(chunks, ignore_index=True)
```
