```python
X = np.array([
    [23, 0.5, 1],
    [35, 1, 1],
    [18, 0, 1]
])

Y = np.array([55, 100, 45])
```

```python
xxt = np.dot(X.T, X)
```

```python
xxt_inv = np.linalg.inv(xxt)
```

```python
xxt_inv
```

```python
xxt_inv_xt = np.dot(xxt_inv, X.T)
```

```python
final_betas = np.dot(xxt_inv_xt, Y)
```
